## 1. Data Types

## 2. Lists and Basic Operations

Lists are ordered, mutable collections. You can add, remove, and modify elements.

````python
# Creating a list
fruits = ["apple", "banana", "orange"]        
numbers = [1, 2, 3, 4, 5]

# Accessing elements (0-based indexing)
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")
print(f"Second number: {numbers[1]}")  

# List length
print(f"\nNumber of fruits: {len(fruits)}")

# Adding elements
fruits.append("grape")
print(f"After adding grape: {fruits}")

# Removing elements
fruits.remove("banana")
print(f"After removing banana: {fruits}")

# List slicing
print(f"First two fruits: {fruits[0:2]}")
print(f"Last two fruits: {fruits[-2:]}")
````

## 3. Dictionaries

Dictionaries store key-value pairs. They're perfect for structured data.

````python
# Creating a dictionary
student = {
    "name": "Bob",
    "age": 20,
    "major": "Computer Science",
    "gpa": 3.8
}

# Accessing values
print(f"Student name: {student['name']}")
print(f"Student age: {student['age']}")

# Safe access with .get()
phone = student.get("phone", "No phone on file")
print(f"Phone: {phone}")

# Adding new key-value pairs
student["phone"] = "555-1234"
print(f"\nUpdated student: {student}")

# Getting all keys and values
print(f"\nKeys: {student.keys()}")
print(f"Values: {student.values()}")
````

## 5. Loops

Use loops to repeat code multiple times.

````python
# For loop
print("For loop - Counting:")
for i in range(1, 7):  # Counts from 1 to 6
    print(f"Count: {i}")

# For loop with a list
print("\nFor loop - Iterating through list:")
colors = ["red", "blue", "green"]
for color in colors:    
    print(f"Color: {color}")

# For loop with enumerate
print("\nFor loop with index:")
for index, color in enumerate(colors):
    print(f"{index}: {color}")
````

````python
# While loop
print("While loop - Countdown:")
count = 5
while count > 0:
    print(f"Count: {count}")
    count -= 2
    
print("Blastoff!")
````

## 6. Functions

Functions are reusable blocks of code.

````python
# Basic function
def greet(name):
    """A simple greeting function."""
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))

# Function with multiple parameters
def add(a, b):
    """Add two numbers."""
    return a + b

result = add(5, 3)
print(f"\n5 + 3 = {result}")

# Function with default parameters
def power(base, exponent=2):
    """Raise base to the power of exponent."""
    return base ** exponent

print(f"\n2^2 = {power(2)}")
print(f"2^3 = {power(2, 3)}")
````

## 7. List Comprehensions (Bonus)

A concise way to create lists.

````python
# Traditional way
squares = []
for i in range(1, 6):
    squares.append(i ** 2)
print(f"Squares (traditional): {squares}")

# List comprehension way
squares = [i ** 2 for i in range(1, 6)]
print(f"Squares (comprehension): {squares}")

# With condition
even_numbers = [i for i in range(1, 11) if i % 2 == 0]
print(f"Even numbers 1-10: {even_numbers}")
````

````python
# 1. DATA STRUCTURE: Initializing the internship-style dataset
data ={
    'Store_ID': [101, 102, 103, None, 105],
    'Inventory_Level': [5, 3, 8, 2, 7],
    'Revenue': [15000, 8000, 22000, 5000, 18000],
    'Location': ['Urban', 'Rural', 'Urban', 'Rural', 'Suburban']
}
df = pd.DataFrame(data)

````

````python
# 2. FUNCTIONS & PARAMETERS: For reusability in Data Cleansing [3]
# Professionals create functions to execute specific, repeatable operations.
def categorize_performance(rev, target=10000):
    """Categorizes store performance based on a revenue target."""
    # 3. CONTROL FLOW (if/elif/else): For Decision-making logic [1]
    if rev >= target * 2:
        return "High Performer"
    elif rev >= target:
        return "Meeting Target"
    else:
        return "Below Target"

# Applying the function using a parameter (target=1000)
df['Performance_Status'] = df['Revenue'].apply(categorize_performance)



````

````python
# 4. LOOPS (for): For automated Data Processing [10]
# Useful for checking inconsistencies across multiple columns at once.
print("--- Automated Missing Value Check ---")
for column in df.columns:
    missing = df[column].isnull().sum()
    if missing > 0:
        print(f"Alert: {column} has {missing} missing values.")
    else:
        print(f"Column '{column}' is clean.")
````
